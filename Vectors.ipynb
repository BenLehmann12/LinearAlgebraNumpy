```python
import numpy as np
import numpy.linalg as lg
```

```python
x,y = np.array([[-2],[2]]),np.array([[3],[3]])
distance = lg.norm(x-y,2)
l1_distance = lg.norm(x-y,1)
print("L2 Distance: ",distance)  #Use Euclidean Distance (L2)
print("L1 Distance: ",l1_distance)  #Use Manhattan Distance (L1)
```


Finding the Angle between 2 Vectors

```python
#The Angle between Two vectors
from numpy import dot, arccos
v = np.array([[5,9,2]])
w = np.array([[6,3,4]])
angle = arccos(dot(v,w.T)/(lg.norm(v)*lg.norm(w)))
print("Angle: ", angle)
```
