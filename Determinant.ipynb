```python
import numpy as np
from numpy.linalg import det
```


Let us say we have a 3x3 Matrix



To find the Determinant of that Matrix, we would solve it like this!

```python
#Simple in Numpy
m = np.array([
    [5,2,1],
    [3,2,7],
    [1,0,2]
])
print('Matrix:\n', m)
print('\nThe Determinant of Matrix M is: %.1f'%det(m))
```

We can produce a 3x3 Identity Matrix with np.eye()

```python
identity = np.eye(3)
print('Identity Matrix:\n', identity)
print('\n M*I:\n', np.dot(m,identity))
```
