```python
import numpy as np
```

Def: Rank is the dimension of the vector space generated by its columns.

```python
def Rank():
    matrixA = np.array([[1, 2, 3, 28],
                       [4, 5, 6, 21],
                       [7, 8, 9, 32],
                       [10, 11, 12, 45]])
    print("The Rank of a Matrix: ", np.linalg.matrix_rank(matrixA))

def main():
    Rank()

if __name__=="__main__":
    main()
```
